# Exploratory Data Analysis

# Univariate Analysis

### Univariate analysis focuses on examining each variable individually. This type of analysis helps in understanding the distribution, central tendency, and dispersion of the data for a single variable. 

```python
print(df.columns)
```

```python
# Summary statistics for the DataFrame
df.describe()
```

```python
df.hist(figsize=(20,20))
# display histogram
plt.show()
```

```python
# Histogram of a numerical column
df['state'].hist()
plt.title('Histogram of State')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()

```

```python
# Box plot of a numerical column
sns.boxplot(x=df['response'])
plt.title('Box Plot of Response')
plt.show()


```

```python
# Frequency count for a categorical column
print(df['state'].value_counts())

```

```python
# Count plot for a categorical column
sns.countplot(x='state', data=df)
plt.title('Count Plot of State')
plt.show()
```

```python
# List of numerical variables
numerical_variables = ['week', 'state', 'campaign']

# Creating box plots for numerical variables
for variable in numerical_variables:
    df.boxplot(column=variable)
    plt.title(f'Box plot of {variable}')
    plt.ylabel(variable)
    plt.show()
```

# Bivariate Analysis

### Bivariate analysis examines the relationship between two variables. This helps in understanding how one variable may influence another and identifying any patterns or correlations. Key methods include:

```python
# Scatter plot between two numerical columns
sns.scatterplot(x='state', y='response', data=df)
plt.title('Scatter Plot between State and Response')
plt.show()
```

```python
#  Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['state'], df['campaign'])
plt.title('Scatter Plot: state vs. campaign')
plt.xlabel('state')
plt.ylabel('campaign')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr()
print(corr_matrix)

```

```python
# Box plot between a categorical column and a numerical column
sns.boxplot(x='state', y='response', data=df)
plt.title('Box Plot of State vs. Response')
plt.show()

```

# Multivariate Analysis

### Multivariate analysis involves examining more than two variables simultaneously to understand complex relationships and interactions among them. This type of analysis can reveal deeper insights that are not apparent from univariate or bivariate analyses alone. Key methods include:

```python
# Bar plot to compare categorical data with numerical data (e.g., mean)
sns.barplot(x='state', y='response', data=df)
plt.title('Bar Plot of State vs. Response')
plt.show()
```

```python
# Pair plot for numerical columns
sns.pairplot(df)
plt.title('Pair Plot')
plt.show()
```

```python
# Enhanced heatmap with more details
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Enhanced Correlation Matrix')
plt.show()
```

```python
# Correlation Matrix
corr_matrix = df[['week', 'state', 'campaign']].corr()
print("Correlation Matrix:")
print(corr_matrix)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Box plot for three variables
sns.boxplot(x='state', y='response', hue='sex', data=df)
plt.title('Grouped Box Plot')
plt.show()
```

```python
# Violin plot to visualize the distribution of numerical data across different categories
sns.violinplot(x='state', y='response', hue='sex', data=df, split=True)
plt.title('Violin Plot')
plt.show()
```

```python
# Pair plot for numerical variables with hue based on 'Sex'
sns.pairplot(df[['week', 'state', 'campaign', 'sex']], hue='sex')
plt.suptitle('Pair Plot of Numerical Variables by Sex', y=1.02)
plt.show()
```
